```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X=pandas.DataFrame(data, columns=['production_budget_usd'])
Y=pandas.DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6)) #width,height
plt.scatter(X,Y,alpha=0.3)  #alpha adds transparency
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0,450000000) #shows the 300000000 units in x-axis
plt.ylim(0,3000000000) #shows the 300000000 units in y-axis
plt.show()
```

```python
regression=LinearRegression() #find the line where th square of the residual sum is minimum
regression.fit(X,Y)
```

```python
regression.coef_ #ouputs the coefficient of the graph
```

```python
#Intercept
regression.intercept_
```

```python
plt.figure(figsize=(10,6)) #width,height
plt.scatter(X,Y,alpha=0.3)  #alpha adds transparency
plt.plot(X,regression.predict(X), color='red', linewidth=4) #gives a lines on the (x, corresponding y)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide Gross $')
plt.xlim(0,450000000) #shows the 300000000 units in x-axis
plt.ylim(0,3000000000) #shows the 300000000 units in y-axis
plt.show()
```

```python
regression.score(X,Y)
```
